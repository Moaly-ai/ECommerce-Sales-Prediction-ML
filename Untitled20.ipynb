```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Month': ['Jan-2025', 'Feb-2025', 'Mar-2025', 'Apr-2025', 'May-2025', 'Jun-2025',
              'Jul-2025', 'Aug-2025', 'Sep-2025', 'Oct-2025', 'Nov-2025', 'Dec-2025'],
    'Marketing_Spend': [5000, 5500, 6000, 6500, 7000, 7500, 8000, 8500, 9000, 9500, 10000, 11000],
    'Website_Visitors': [12000, 12800, 13500, 14200, 15000, 15800, 16500, 17100, 17900, 18600, 19400, 21000],
    'Orders': [320, 350, 380, 410, 440, 470, 500, 520, 550, 580, 620, 680],
    'Avg_Product_Price': [45, 47, 46, 48, 49, 50, 51, 52, 53, 54, 55, 57],
    'Sales': [14400, 16450, 17480, 19680, 21560, 23500, 25500, 27040, 29150, 31320, 34100, 38760]
}
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# 1. تحديد المتغيرات المستقلة (X) والمتغير المستهدف التنبؤ به (y)
X = df[['Marketing_Spend', 'Website_Visitors', 'Orders', 'Avg_Product_Price']]
y = df['Sales']
```

```python
# 2. تقسيم البيانات لتدريب واختبار (80% تدريب و 20% اختبار)
# استخدمنا shuffle=False لأنها بيانات زمنية مرتبطة بالأشهر
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
# 3. بناء وتدريب نموذج Linear Regression
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
lr_pred = lr_model.predict(X_test)
```

```python
# 4. بناء وتدريب نموذج Random Forest
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
```

```python
# 5. حساب مقاييس الأداء لنموذج Linear Regression
lr_mae = mean_absolute_error(y_test, lr_pred)
lr_rmse = np.sqrt(mean_squared_error(y_test, lr_pred))
lr_r2 = r2_score(y_test, lr_pred)
```

```python
# 6. حساب مقاييس الأداء لنموذج Random Forest
rf_mae = mean_absolute_error(y_test, rf_pred)
rf_rmse = np.sqrt(mean_squared_error(y_test, rf_pred))
rf_r2 = r2_score(y_test, rf_pred)
```

```python
# 7. عرض مقارنة النتائج في جدول منظم
results = pd.DataFrame({
    'Metric': ['MAE', 'RMSE', 'R² Score'],
    'Linear Regression': [lr_mae, lr_rmse, lr_r2],
    'Random Forest': [rf_mae, rf_rmse, rf_r2]
})

results
```
